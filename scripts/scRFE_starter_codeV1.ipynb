```python
# make getLabels
def getLabelsV1 (dataMatrix, classOfInterest):
    labels = list(set(dataMatrix.obs[classOfInterest]))
    labelsDict = {}
    lists = [[] for i in range(len(labels))]
    for i in range (0, len(lists)):
        for obs in dataMatrix.obs[classOfInterest]:
            if obs == label:
                lists[i].append('TRUE')
            else:
                lists[i].append('false')
            labelsDict[obs] = lists[i]
    return labelsDict
```
